### RFM Analysis

RFM analysis is a marketing technique used to analyze customer behavior based on three factors: Recency, Frequency, and Monetary value of purchases. It helps businesses identify their most valuable customers and design targeted marketing strategies.

We segment the customers into groups of certain similarities. Then we can begin to answer some business questions

```python
rfm_df = data[[ 'rfm_Recency', 'rfm_Frequency',
       'rfm_MonetaryValue']]
rfm_df.head()
```

```python
# scale the dataframe
scaler = StandardScaler()
scaled_df = scaler.fit_transform(rfm_df)
```

First, we run a range of different K-values to find the optimal K

```python
inertia = []
val_range = range(1,11)
for i in val_range:
    kmean = KMeans(n_clusters=i)
    kmean.fit(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)

inertia
```

```python
plt.plot(val_range,inertia,'bo-')
plt.xlabel('K-Values') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method') 
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)           
kmeans.fit_predict(scaled_df)         
labels = kmeans.labels_
kmeans.inertia_
```

Let's plot view the clusters using PCA. We use PCA to reduce the number of dimensions,i.e, the number of columns down to two.

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

```python
pca_df["cluster"] = labels
pca_df
```

```python
# plot the clusters
plt.figure(figsize=(8,4))
ax = sns.scatterplot(x="PCA1", y="PCA2", hue="cluster", data=pca_df, \
                     palette=['red', 'green', 'blue', 'black'], s=50)
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

```python
# add the the labels to rfm_df to see each customer and their corresponding label/group.
clustered_df = rfm_df.copy()
clustered_df["cluster"] = labels
```

```python
def get_status(row):
    if row["cluster"] == 0:
        return "Loyal"
    elif row["cluster"] == 1:
        return "hybernating/Loosing"
    elif row["cluster"] == 2:
        return "Lost"
    else:
        return "New"
      
clustered_df["Cus_status"] = clustered_df.apply(get_status,axis=1)
```

# Clustering Result Overview

```python
# check the mean of each feature for each group
clustered_df.groupby("cluster").mean()
```

```python
# Check the the new dataframe with the clusters 
clustered_df = clustered_df.reset_index()
clustered_df.head(3)
```

```python
# number of customers in each cluster
clustered_df["Cus_status"].value_counts()
```

```python
sns.countplot(x='Cus_status', data=clustered_df)
plt.show()
```

**Cluster_0:Loyal**

Cluster_0 comprises regular customers with high recency, frequency, and monetary value, who contribute significantly to the overall sales amount. They can be considered our **Loyal Customers**. Offering rewards through a loyalty program, such as exclusive discounts, free gifts, or early access to new products, can make them feel appreciated and incentivize them to continue shopping with your brand.

**Cluster_1:Hybernating/Losing**

Customers in Cluster_1 have a high potential for churn since they have low recency scores, indicating they haven't made a purchase in a while. However, since they have high frequency and monetary value, they present an opportunity for increased revenue if we can retain them. To prevent losing these valuable customers, we can implement strategies such as personalized promotions, product recommendations, email reminders, and loyalty rewards to incentivize them to make a purchase and continue their patronage.

**Cluster_2:Lost/Losing**

Cluster_2 includes customers who haven't made a recent purchase, have a low frequency of buying, and low monetary value. These customers are categorized as **lost or hibernating customers**. To bring them back, a targeted re-engagement campaign can be launched to remind them of the brand and entice them to make a purchase through incentives or special promotions. This strategy can help reignite their interest and bring them back to being active customers.

**_Cluster_3:New**

Cluster_3 can be considered **new customers** who have a very high recency score, indicating that they recently made a purchase. However, their purchase frequency and monetary value are relatively low, which is expected for new customers. To encourage them to become repeat customers, we can offer them welcome offers, such as discounts on their next purchase, free shipping, or a free gift. This approach can incentivize them to make a second purchase and increase the likelihood of them becoming loyal customers.
